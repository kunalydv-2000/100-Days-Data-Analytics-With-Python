# common string methods

```python
# upper()

name = "python"

print(name.upper())
```

```python
# lower()

name = "PYTHON"

print(name.lower())
```

```python
# title()

course = "dAtA AnalyticS"

print(course.title())
```

```python
# capitalize()

city = "deLHi metro"

print(city.capitalize())
```

```python
# replace()

sentence = "I like Java"

print(sentence.replace("Java", "Python"))
```

```python
# strip()

name = "   Rahul   "

print(name.strip())
```

```python
# string concatenation

first_name = "Rahul"
last_name = "Sharma"

full_name = first_name + " " + last_name

print(full_name)
```

```python
# string repetition

print("Python " * 3)
```

```python
# membership operators

course = "Data Analytics"

print("Data" in course)
print("Python" in course)
```

```python
# string formatting

name = "Rahul"
age = 25

print(f"My name is {name}")
print(f"I am {age} years old")
```

```python
# escape characters

print("Python\nData Analytics")
```

```python
print("Name\tAge")
```

```python
# count characters

name = "banana"

print(name.count("a"))
```

```python
# find position

name = "Analytics"

print(name.find("t"))
```

```python
# starts with

email = "user@gmail.com"

print(email.startswith("user"))
```

```python
# ends with

email = "user@gmail.com"

print(email.endswith(".com"))
```
