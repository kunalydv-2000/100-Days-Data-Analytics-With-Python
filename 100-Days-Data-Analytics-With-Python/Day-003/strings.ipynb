# What is a String?
A string is a sequence of characters enclosed in quotes.

```python
name = "Rahul"
city = 'Delhi'
course = "Data Analytics"
```

# String Indexing
Every character has a position (index).

```python
name = "Python"

print(name[0])
print(name[1])
print(name[5])
```

# Negative Indexing
Python can access characters from the end.

```python
print(name[-1])
print(name[-2])
```
