# STRING TO INTEGER

```python
print("STRING TO INTEGER")

age_str = "25"

age_int = int(age_str)

print(age_str)
print(type(age_str))

print(age_int)
print(type(age_int))
```

# INTEGER TO STRING

```python
print("INTEGER TO STRING")

score = 95

score_str = str(score)

print(score)
print(type(score))

print(score_str)
print(type(score_str))
```

# INTEGER TO FLOAT

```python
print("INTEGER TO FLOAT")

number = 100

number_float = float(number)

print(number)
print(type(number))

print(number_float)
print(type(number_float))
```

# FLOAT TO INTEGER

```python
print("FLOAT TO INTEGER")

price = 99.99

price_int = int(price)

print(price)
print(type(price))

print(price_int)
print(type(price_int))
```

# STRING TO FLOAT

```python
print("STRING TO FLOAT")

salary = "50000.50"

salary_float = float(salary)

print(salary)
print(type(salary))

print(salary_float)
```

# BOOLEAN CONVERSION

```python
print("BOOLEAN CONVERSION")

print(bool(1))
print(bool(0))
print(bool("Python"))
print(bool(""))
```

# PRACTICAL EXAMPLE

```python
print("PRACTICAL EXAMPLE")

age = int("30")
salary = float("45000.75")

print(f"Age: {age}")
print(f"Salary: {salary}")
```
