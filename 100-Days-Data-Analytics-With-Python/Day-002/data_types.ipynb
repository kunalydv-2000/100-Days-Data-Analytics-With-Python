# **Data Types**

```python
# string (str)
name = 'kunal'
city = 'gurugram'

print('String examples')
print(name)
print(city)
```

```python
# integer (int)
age = 25
year = 2026

print('integer example')
print(age)
print(year)
```

```python
# float (float)
height = 5.7
salary = 45000.50

print('float example')
print(height)
print(salary)
```

```python
# boolean (bool)
is_student = True
is_employee = False

print('boolean example')
print(is_student)
print(is_employee)
```

```python
# type checking
print('type checking')

print(type("Python"))
print(type(100))
print(type(99.99))
print(type(True))
```

```python
print("DATA TYPE SUMMARY")

print(f"Name: {name} -> {type(name)}")
print(f"Age: {age} -> {type(age)}")
print(f"Height: {height} -> {type(height)}")
print(f"Student: {is_student} -> {type(is_student)}")
```
